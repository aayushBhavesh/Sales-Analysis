Total Revenue vs Discount Impact by Category Scenario: The Marketing Team wants to analyze the total revenue for each product category before and after applying discounts. They also want to visualize the discount impact on revenue across different categories. Requirements: â€¢ Create a bar chart showing total revenue (before and after discount) for each product category. â€¢ Use the formula SUM(before_discount) for total revenue before discount and SUM(after_discount) for total revenue after discount. â€¢ Add a calculated field to show the discount impact (before_discount - after_discount). â€¢ Visualize the difference in revenue for each category. Key Features to Use: â€¢ category â€¢ before_discount â€¢ after_discount

```python
# ðŸ“Œ Task: Total Revenue vs Discount Impact by Category

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
file_path = r"C:\Users\bhave\OneDrive\Desktop\dashboard\dataset2\finaldataset.csv"
df = pd.read_csv(file_path)

# Step 2: Group by category and calculate revenues
category_revenue = df.groupby("category", as_index=False).agg({
    "before_discount": "sum",
    "after_discount": "sum"
})

# Step 3: Calculate discount impact
category_revenue["discount_impact"] = category_revenue["before_discount"] - category_revenue["after_discount"]

# Step 4: Display results
print("ðŸ“Š Revenue Analysis by Category:")
print(category_revenue)

# Step 5: Visualization
plt.figure(figsize=(10,6))

# Bar chart for before & after discount
bar_width = 0.35
categories = category_revenue["category"]
x = range(len(categories))

plt.bar(x, category_revenue["before_discount"], width=bar_width, label="Before Discount", alpha=0.8)
plt.bar([i + bar_width for i in x], category_revenue["after_discount"], width=bar_width, label="After Discount", alpha=0.8)

# Add discount impact as red line (difference)
plt.plot([i + bar_width/2 for i in x], category_revenue["discount_impact"], 
         color="red", marker="o", linestyle="--", label="Discount Impact")

plt.xticks([i + bar_width/2 for i in x], categories, rotation=45)
plt.ylabel("Revenue")
plt.title("Total Revenue vs Discount Impact by Category")
plt.legend()
plt.tight_layout()
plt.show()
```
