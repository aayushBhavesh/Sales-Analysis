Average Quantity Sold per Product by Category Scenario: The Product Team wants to know the average quantity sold for each product category to identify potential inventory issues. Requirements: â€¢ Create a bar chart that shows the average quantity sold (AVG(qty_ordered)) for each product category. â€¢ Sort the categories by average quantity sold in descending order. Key Features to Use: â€¢ category â€¢ qty_ordered

```python
# ðŸ“Œ Task: Average Quantity Sold per Product by Category

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
file_path = r"C:\Users\bhave\OneDrive\Desktop\dashboard\dataset2\finaldataset.csv"
df = pd.read_csv(file_path)

# Step 2: Group by category and calculate average qty_ordered
avg_qty = (
    df.groupby("category", as_index=False)["qty_ordered"]
    .mean()
    .sort_values(by="qty_ordered", ascending=False)
)

# Step 3: Display result
print("ðŸ“Š Average Quantity Sold per Category:")
print(avg_qty)

# Step 4: Visualization
plt.figure(figsize=(10,6))
plt.bar(avg_qty["category"], avg_qty["qty_ordered"], color="skyblue")
plt.title("Average Quantity Sold per Product by Category")
plt.xlabel("Category")
plt.ylabel("Average Quantity Sold")
plt.xticks(rotation=45, ha="right")
plt.show()
```
